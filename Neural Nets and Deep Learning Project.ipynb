# Part 1. Import Package(s) and Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('tensorflow_regression_df.csv')

df
```

# Part 2. Define X and Y Values and Split Train and Test Data

```python
X = df[['Shots pg','Dribbles pg']].values

y = df['Goals Conceded'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
scaler = MinMaxScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)

X_test = scaler.transform(X_test)
```

# Part 3. Build Deep Learning Model to Predict the Number of Goals Conceded

```python
model = Sequential([Dense(4, activation = 'relu'),
                   Dense(2,activation='relu'),
                   Dense(1)])

model = Sequential()

model.add(Dense(4, activation = 'relu'))
model.add(Dense(2, activation = 'relu'))
model.add(Dense(1, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'rmsprop', loss = 'mse')
```

```python
model.fit(x = X_train, y = y_train, epochs = 250)
```

```python
pd.DataFrame(model.history.history)
```

```python
model.evaluate(X_test, y_test, verbose = 0)
```

```python
test_predictions = model.predict(X_test)

test_predictions
```

```python
test_predictions = pd.Series(test_predictions.reshape(57,))

pred_df = pd.DataFrame(y_test, columns = ['Test True Y'])

pred_df = pd.concat([pred_df, test_predictions], axis = 1)

pred_df
```

# Part 4. Evaluate Performance of Model using Mean Absolute Error

```python
mean_absolute_error(pred_df['Test True Y'], pred_df['Model Predictions'])
```
